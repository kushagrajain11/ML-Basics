```python
from sklearn import datasets
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import os
import re
```

```python
path = "/home/venom/Desktop/Untitled Folder/20_newsgroups/"
os.chdir(path)
folder = os.listdir(path)
folder
```

```python
folder
num_categories = 20
num_articles = 980
word_dict = {}
```

```python
for i in range(num_categories):
    curr_folder = path + folder[i]
    articles = os.listdir(curr_folder)
    for j in range(num_articles):
        curr_article = curr_folder +'/'+ articles[j]
        with open(curr_article, 'r', errors='ignore') as f:
            for line in f:
                for word in re.findall(r'\w+', line):
                    word_dict[word]=1 if not word_dict.get(word) else word_dict[word]+1
```
