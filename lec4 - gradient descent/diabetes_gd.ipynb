```python
from sklearn import datasets
import pandas as pd
import numpy as np
from sklearn import cross_validation as cv
from sklearn.model_selection import train_test_split
```

```python
diabetes=datasets.load_boston()
df=pd.DataFrame(diabetes.data)
df.columns = diabetes.feature_names
df.loc[:, 'NewCol'] = 1
df.head()
x_train, x_test, y_train, y_test= train_test_split(df, diabetes.target, test_size=0.2)
```

```python
def step_gradient(m, alpha):
    slopes=np.zeros(len(m))
    n=len(m)
    for i in range(len(x_train.values)):
        for j in range(n):
            x=x_train.values[i]
            y=y_train[i]
            slopes[j]+= -(2/n) * (y-x.dot(m))*x[j]
    new_m=np.zeros(n)
    new_m=m-alpha*slopes
    return new_m

def cost_gd(m):
    total_cost = 0
    for i in range(len(x_train.values)):
        x = x_train.values[i]
        y = y_train[i]
        total_cost += (y - m.dot(x))**2
    return total_cost/len(x_train.values)

def gd_runner(alpha = 0.00000011, num_iterations = 10000):
    m = np.zeros(len(df.columns))
    for i in range(num_iterations):
        print(cost_gd(m))
        m = step_gradient(m, alpha)
    print ("FINAL: ", cost_gd(m))
    return (m)
```

```python
m=gd_runner()
```

```python
def predict(x_test, m):
    y_pred=np.zeros(len(x_test.values))
    for i in range(len(x_test.values)):
        y_pred[i]=x_test.values[i].dot(m)
    return y_pred

y_pred = predict(x_test, m)
```

```python
print(np.sum((y_pred - y_test)**2))
print(y_pred)
print(y_test)
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['legend.numpoints'] = 1

#generate some random data
pred = y_pred
GT =  y_test

fig, ax = plt.subplots(figsize=(20,7))

# plot a black line between the 
# ith prediction and the ith ground truth 
for i in range(len(pred)):
    ax.plot([i,i],[pred[i], GT[i]], c="k", linewidth=0.5)
ax.plot(pred,'o', label='Prediction', color = 'g')
ax.plot(GT,'^', label='Ground Truth', color = 'r' )

ax.set_xlim((-1,100))
plt.xlabel('a')
plt.ylabel('b')
plt.title('test')

plt.legend()             
plt.show()
```
